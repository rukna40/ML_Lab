```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot([0,100],[0,200])
ax.set_xticks(range(0,101,20))
ax.set_yticks(range(0,201,50))
plt.xlim(0,100)
plt.ylim(0,200)
plt.xlabel("x")
plt.ylabel("y")
plt.title('title')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot([0,100],[0,100])
ax2.plot([0,10],[0,10])
plt.show()
```

```python
df=pd.read_csv('company_sales_data.csv')
plt.plot(df['month_number'],df['total_profit'])
plt.xticks(df['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.plot(df['month_number'],df['total_units'],linestyle='dotted',color='red',label='line',marker='o',linewidth=3)
plt.xticks(df['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.show()
```

```python
df1=df.iloc[:,[1,2,3,4,5,6]]
plt.plot(df['month_number'],df1)
plt.xticks(df['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(df1.columns, loc='upper left')
plt.show()
```

```python
plt.pie(df1.sum(),labels=df1.columns,autopct='%.2f%%')
plt.show()
```
